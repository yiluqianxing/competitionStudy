```python
# 没有重复的表全部拼接起来
train = pd.read_csv("../input/train_auth_info.csv")
test = pd.read_csv("../input/test_auth_info.csv")

train = train.merge(pd.read_csv("../input/train_credit_info.csv"),how = "inner",on = "id")
test = test.merge(pd.read_csv("../input/test_credit_info.csv"),how = "inner",on = "id")

mm = pd.read_csv("../input/train_user_info.csv")
nn = pd.read_csv("../input/test_user_info.csv")
mm = mm[["sex","qq_bound","wechat_bound","id"]]
nn = nn[["sex","qq_bound","wechat_bound","id"]]

train = train.merge(mm,how = "inner",on = "id")
test = test.merge(nn,how = "inner",on = "id")

train.head(),test.head(),train.shape,test.shape
```

```python
import time
def gettime(x):
    t = time.strptime(x, "%Y-%m-%d")  #秒的话加上秒
    t = int(time.mktime(t))
    return t
def gettime8(x):
    x = x.split(".")[0]
    t = time.strptime(x, '%Y-%m-%d %H:%M:%S')  #秒的话加上秒
    t = int(time.mktime(t))
    return t
def myfillna(fulldata,lab):
    fulldata[lab].fillna(fulldata[lab].value_counts().sort_values(ascending = False).index[0],inplace = True)
    
def getlast4(x):
    tmp = x[len(x)-4:]
    if tmp[0] == "*":
        return 3
    else:
        return 4
```

```python
#对合并的单表进行特征工程
import numpy as np
fulldata = pd.concat([train,test],axis = 0,ignore_index = True)
# print(fulldata.head(),"\n",fulldata.shape)
fulldata["missNumber"] = fulldata.isnull().sum(axis = 1)
label = fulldata.columns 
more1 = fulldata["auth_time"].value_counts().sort_values(ascending = False).index[0]
for i in range(fulldata.shape[0]):
    if (fulldata["auth_time"][i] == '0000-00-00'):
        fulldata["auth_time"][i] = more1

#同列的缺失率
for i in range(fulldata.shape[1]):
    print("以列为单位进行统计")
    print("feature:",label[i])
    print("缺失率 = ",fulldata[label[i]].isnull().sum()/fulldata.shape[0]) 
    if fulldata[label[i]].isnull().any():
        myfillna(fulldata,label[i])

# import numpy as np
# np.max(train.isnull().sum(axis = 1)) ,np.max(test.isnull().sum(axis = 1)) #1横着走#1横着走

fulldata["auth_time"].value_counts().sort_values(ascending = False)
print(fulldata["auth_time"].isnull().sum())
fulldata["auth_time"] = fulldata["auth_time"].astype("str")

fulldata["auth_time_miao"] = fulldata["auth_time"].apply(lambda x:gettime(x))
fulldata["auth_time_year"] = fulldata["auth_time"].apply(lambda x:x.split("-")[0])
fulldata["auth_time_mon"] = fulldata["auth_time"].apply(lambda x:x.split("-")[1])
fulldata["auth_time_day"] = fulldata["auth_time"].apply(lambda x:x.split("-")[2])
fulldata["phone"] = fulldata["phone"].astype("str")
fulldata["phoneTop3"] = fulldata["phone"].apply(lambda x:x[0:3])
fulldata["phoneLastLen"] = fulldata["phone"].apply(lambda x:getlast4(x))
sex = pd.get_dummies(fulldata.sex,prefix = "sex")
qq_bound = pd.get_dummies(fulldata['qq_bound'],prefix = 'qq_bound')
wechat_bound = pd.get_dummies(fulldata['wechat_bound'],prefix = 'wechat_bound')
fulldata = pd.concat([fulldata,sex,qq_bound,wechat_bound],axis = 1)
```

```python
print(fulldata.columns,fulldata.shape)
# fulldata["phone"] = fulldata["phone"].astype("str")
# fulldata["phone"].apply(lambda x:x[0:3])
train = fulldata.iloc[0:train.shape[0],:]
test = fulldata.iloc[train.shape[0]:,:]
trainTarget = pd.read_csv("../input/train_target.csv")
testTarget = pd.read_csv("../input/test_list.csv")
trainTarget["trainAppl_sbm_tm"] =  trainTarget.appl_sbm_tm.apply(lambda x:gettime8(x))
testTarget["testAppl_sbm_tm"] =  testTarget.appl_sbm_tm.apply(lambda x:gettime8(x))
# 先拼接起来保险，不然容易错
train = train.merge(trainTarget,how = "left",on = "id")
test = test.merge(testTarget,how = "left",on = "id")
train["timeGap"] = train["trainAppl_sbm_tm"] - train['auth_time_miao'] 
test["timeGap"] = test["testAppl_sbm_tm"] - test['auth_time_miao'] 
# print(len(test["timeGap"]),len(testAppl_sbm_tm))
# testAppl_sbm_tm.head(),test['auth_time_miao']
print(train.isnull().sum(),test.isnull().sum())
print(train.head(),test.head())

del train["id_card"]
del test["id_card"]
del train["auth_time"]
del test["auth_time"]
del train["phone"]
del test["phone"]
del train["sex"]
del test["sex"]
del train["qq_bound"]
del test["qq_bound"]
del train["wechat_bound"]
del test["wechat_bound"]
del train["appl_sbm_tm"]
del test["appl_sbm_tm"]
# train.head(),train.shape,'\n',test.head(),test.shape
```

```python
m = train.target.copy()
del train["target"]
train["target"] = m
train.head(),test.head(),'\n',train.shape,test.shape,train.columns,test.columns
```
