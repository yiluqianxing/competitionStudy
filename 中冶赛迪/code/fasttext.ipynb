```python
#生成文本格式
import jieba
import pandas as pd
import random
```

```python
train =  pd.read_csv("../trainFullNumber.csv",encoding = "utf-8",header = None)
# validata =  pd.read_csv("../validataFull.csv",encoding = "utf-8",header = None)
train = train.iloc[:,0:2]
train
train.columns = ["Y","content"]
# validata.columns = ["Y","content"]
# train.head(),validata.head(),train.shape,validata.shape
```

```python
#不用停用词
def prepross_text(content_lines,sentences,category):
    number = 0
    for line in content_lines:
        try:
            segs = jieba.cut(line)
            sgs = filter(lambda x: x != ' ',segs)
            sentences.append("_label_"+str(category.iloc[number])+" , "+" ".join(segs))
            number += 1
        except (Exception,e):
            print (line)
            continue 
```

```python
#生成训练数据
sentences = []
prepross_text(train.content,sentences,train.Y)
```

```python
sentences
```

```python
import fasttext
import random
random.shuffle(sentences)
# 'train_data.txt' 将上面生成的格式保存下来
print ("writing data to fasttext format...")
out = open('train_data.txt', 'w')
for sentence in sentences:
    out.write(str(sentence.encode('utf8')+b"\n"))
    # 这是python3 的做法
print ("done!")
```

```python
classifier = fasttext.supervised("train_data.txt","classifier.model",label_prefix = '_label_')
result = classifier.test("train_data.txt")
print("precision:",result.precision)
print("recall:",result.recall)
```
